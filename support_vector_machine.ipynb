```python
# fig, ax = plt.subplots()
# # print(aaron_judge.columns)
# # print(aaron_judge.description.unique())
# # print(aaron_judge.type.unique())

# aaron_judge['type'] = aaron_judge['type'].map({'S':1, 'B':0})
# # print(aaron_judge.type)
# # print(aaron_judge['plate_x'])

# aaron_judge.dropna(subset = ['plate_x', 'plate_z', 'type'], inplace=True)
# # print(aaron_judge.plate_x.unique())
# # print(aaron_judge.plate_z.unique())
# # print(aaron_judge.type.unique())

# # plt.scatter(aaron_judge['plate_x'], aaron_judge['plate_z'], c = 'purple', cmap = plt.cm.coolwarm, alpha = 0.25)

# training_set, testing_set = train_test_split(aaron_judge, random_state = 1)

# # classifier = SVC(kernel='rbf')
# # # classifier = SVC(kernel='rbf', gamma = 100, C = 100)
# # # classifier = SVC(kernel='rbf', gamma = 3, C = 1)
# # classifier.fit(training_set[['plate_x', 'plate_z']], training_set['type'])

# # print(classifier.score(testing_set[['plate_x', 'plate_z']], testing_set['type']))

# params = range(1,10)
# gamma, c = 1, 1
# best_params = [gamma, c]
# best = 0

# for i in params:
#   for j in params:
#     classifier = SVC(kernel='rbf', gamma = i, C = j)
#     classifier.fit(training_set[['plate_x', 'plate_z']], training_set['type'])
#     score = classifier.score(testing_set[['plate_x', 'plate_z']], testing_set['type'])
#     if score > best:
#       best_params = [i, j]
#       best = score

# print('best parameters', best_params)
# classifier = SVC(kernel='rbf', gamma = best_params[0], C = best_params[1])
# classifier.fit(training_set[['plate_x', 'plate_z']], training_set['type'])
# print('score', classifier.score(testing_set[['plate_x', 'plate_z']], testing_set['type']))

# ax.set_ylim(-2, 2)
# draw_boundary(ax, classifier)

# plt.show()

def predScore(player, add_strikes=False):

    player['type'] = player['type'].map({'S':1, 'B':0})
    player.dropna(subset = ['plate_x', 'plate_z', 'type'],     inplace=True)

    plt.scatter(player['plate_x'], player['plate_z'], c =  'purple', cmap = plt.cm.coolwarm, alpha = 0.25)

    training_set, testing_set = train_test_split(player, random_state = 1)
  
    params = range(1,10)
    gamma, c = 1, 1
    best_params = [gamma, c]
    best = 0
    train = training_set[['plate_x', 'plate_z']]
    test = testing_set[['plate_x', 'plate_z']]

    if add_strikes:
        print('With strikes')
        train = training_set[['plate_x', 'plate_z', 'strikes']]
        test = testing_set[['plate_x', 'plate_z', 'strikes']]

    for i in params:
        for j in params:
            classifier = SVC(kernel='rbf', gamma = i, C = j)
            classifier.fit(train, training_set['type'])
            score = classifier.score(test, testing_set['type'])
            if score > best:
                best_params = [i, j]
                best = score

    print('best parameters', best_params)
    classifier = SVC(kernel='rbf', gamma = best_params[0], C = best_params[1])
    classifier.fit(train, training_set['type'])
    print('score', classifier.score(test, testing_set['type']))

print('Aaron Judge')
# predScore(aaron_judge)
predScore(aaron_judge, add_strikes=True)

print('Jose Altuve')  
# predScore(jose_altuve)
predScore(jose_altuve, add_strikes=True)

print('David Ortiz')  
# predScore(david_ortiz)
predScore(david_ortiz, add_strikes=True)
```
